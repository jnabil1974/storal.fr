```python
import requests
import re
from urllib.parse import urljoin

url = 'https://storal.fr/products/kissimy'
print(f"Fetching {url}...")
response = requests.get(url)

print(f"Page Status Code: {response.status_code}")

if response.status_code == 200:
    html_content = response.text
    # Find CSS file using regex since we might not have bs4 installed in the env, but let's try regex first as it's built-in
    # Look for href="/_next/static/css/..."
    css_matches = re.findall(r'href="(/_next/static/css/[^"]+\.css)"', html_content)
    
    if css_matches:
        print(f"Found {len(css_matches)} CSS files.")
        for css_path in css_matches:
            full_css_url = urljoin(url, css_path)
            print(f"Checking CSS: {full_css_url}")
            css_response = requests.get(full_css_url)
            print(f"CSS Status Code: {css_response.status_code}")
            if css_response.status_code == 404:
                print(f"ERROR: CSS File not found: {css_path}")
            else:
                 print("CSS loaded successfully.")
    else:
        print("No CSS files found in HTML matching pattern /_next/static/css/")
        print("First 500 chars of HTML:")
        print(html_content[:500])
else:
    print("Failed to load page.")
```
